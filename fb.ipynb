```python
DF = pd.read_csv('facebookdata.csv')
```

```python
dataframe = DF.iloc[:,1:2]
DF_list = dataframe.values.flatten()

comment_list = list(DF_list)
comment_list
```

```python
import re 
emoji_pattern = re.compile("["
         u"\U0001F600-\U0001F64F"  # emoticons
         u"\U0001F300-\U0001F5FF"  # symbols & pictographs
         u"\U0001F680-\U0001F6FF"  # transport & map symbols
         u"\U0001F1E0-\U0001F1FF"  # flags (iOS)
         u"\U00002702-\U000027B0"
         u"\U000024C2-\U0001F251"              
         "]+", flags=re.UNICODE)
l = emoji_pattern.sub(r'',str(comment_list))
            
```

```python
#z = l.split('\',')
res = l.strip('][').split('\',') 
```

```python
import nltk
from nltk.sentiment.vader import SentimentIntensityAnalyzer
test_subset=['MANISH OPEN UP ::,,,,,','ACHIN IS GOOD!!!!!']

sid = SentimentIntensityAnalyzer()
pos_word_list=[]
neu_word_list=[]
neg_word_list=[]

for word in res:
    if (sid.polarity_scores(word)['compound']) >= 0.5:
        pos_word_list.append(word)
    elif (sid.polarity_scores(word)['compound']) <= -0.5:
        neg_word_list.append(word)
    else:
        neu_word_list.append(word)                

#print('Positive :',pos_word_list)        
#print('Neutral :',neu_word_list)    
print('Negative :',neg_word_list[12]) 
```
