## Importing all the necessary libraries required.

```python

from nltk.corpus import stopwords     # importing nltk.corpus library
import nltk                           # importing nltk package
nltk.download('stopwords')            # downloading stopwords from nltk
from nltk.stem.wordnet import WordNetLemmatizer  # importing WordNetLemmatizer form nltk.stem.wordnet library
import string                        # importing string library

from textblob.classifiers import NaiveBayesClassifier # importing textblob.classifiers
import pandas as pd                   # importing pandas


```

## Defining global variables.

```python
global df
global emotion
global text
global data
global stop_words
global punctuation
global lemmatized
global neg
global nb
```

```python
df = pd.read_csv('data_emotion.csv', header=None)  # importing the dataset from data_emotion.csv file
emotion = pd.Series(df[0])
text = pd.Series(df[1])

# Storing some pre-defined negative words in 'neg'.
neg = ['not', 'nor', 'neither', 'although', 'but', 'however', 'nonetheless', 'even though', 'except','despite', 'yet']
data = pd.DataFrame({'Text': text, 'Emotion': emotion})   #Creating a dictionary of text:emotions.
 

stop_words = set(stopwords.words('english'))  # Storing all stop words from 'english' language.
punctuation= set(string.punctuation)          # Storing all the punctuations.
lemmatized = WordNetLemmatizer()              
list_of_emotions=[]                           # List to store emotions
list_of_text=[]                               # List to store comments
training_data=[]                              # List to store training data

```

## Data Preprocessing.

```python
     #  Selecting  texts for testing data.
test_data_emotions = data['Emotion']                 # Retrieving the 'Emotion' column from the dataset.
test_data_emotions = test_data_emotions[6000:]     # Selecting emotions for testing data.
```

### Creating preprocess() function to remove stopwords, negative words and punctuations from the text.

```python
def preprocess(doc):   
        # Removing stopwords and negative words. Also, converting all the words to lower case.
        stop_excluding = " ".join([i for i in doc.lower().split() if i not in stop_words if i not in neg])
        
        # Removing all punctuations from the above preprocessed data.
        punc_free = "".join([ch for ch in stop_excluding if ch not in punctuation])
        
        # Lemmatizing all the words(the data is free of stopwords,negative words and punctuation)
        normalized = " ".join([lemmatized.lemmatize(word) for word in punc_free.split()])
        normalized = normalized.replace("\r\n","")
        return normalized
```

```python
# Preprocessing all the data rows by calling the preprocess() fuction for each row.
for i in range(df.shape[0]):
    data.loc[i]['Text'] = preprocess(data.loc[i]['Text'])
    
```
